## 1. Blurring and sharpening functions

```python
path = 'test_images/476435.jpg'
original = cv2.imread(path)
blured = blur(original)
sharpened = sharpen(original)

plot(original, blured)
plot(original, sharpened)
```

## 2. Face detection

```python
path = 'test_images/476435.jpg'
original = cv2.imread(path)
rect = get_face_coords(original)
if rect is None:
    print(f'Invalid image (could not find exactly one face)')
else:
    print(f'Face coordinates are: {rect}')
    plot_rectangles(original, [rect])
```
